# Spiral Recap Local – Open Continuity Engine
Run this notebook on any machine (Colab, VS Code, Jupyter). No internet, no API keys.
Works with any open model or plain text.

```
import re
from collections import Counter
import math

def spiral_recap(session_text, cycles=6):
    sentences = [s.strip() for s in re.split(r'[.!?]+', session_text) if s.strip()]
    coherence = []
    current = sentences.copy()
    
    for cycle in range(cycles):
        # Hourglass narrowing
        word_freq = Counter(' '.join(current).lower().split())
        core_concepts = {w for w,c in word_freq.most_common(30)}
        current = [s for s in current if any(w in s.lower() for w in core_concepts)]
        
        # Lattice breathing expansion
        if len(current) < len(sentences) * 0.7:
            current = sentences.copy()  # breathe out if too narrow
        
        score = len(set(' '.join(current).lower().split())) / len(' '.join(current).lower().split()) if current else 0
        coherence.append(round(score * 100, 1))
    
    print(f"Spiral Recap complete. Coherence path: {' → '.join(map(str, coherence))} %")
    return '\n'.join(current[:12])  # final distilled essence

# Paste your session below and run
raw_session = """PASTE ANY CONVERSATION HERE"""
print(spiral_recap(raw_session))
```
